```python
import pandas as pd

pd.set_option('display.max_columns', None)

database = pd.read_csv(r'C:\Users\eduar\OneDrive\Desktop\DEV\Football\PremierLeague\database.csv',index_col=None)
```

```python
# Agrupar os dados por jogador e calcular a soma dos chutes ao gol e a contagem de aparições
db_agrupado_jogador = database.groupby(['Player','Team'], as_index= False).agg(
    Total_Gols=('Goals','sum'),
    Total_Chutes=('Total Shoot', 'sum'),   # Soma dos chutes ao gol
    Aparicoes=('Player', 'count'), # Contagem de aparições
    Total_Chutes_Gol=('Shoot on Target', 'sum') # Contagem de aparições
)
```

```python
db_agrupado_jogador['Média_chute']=round(db_agrupado_jogador['Total_Chutes'] / db_agrupado_jogador['Aparicoes'],2)
db_agrupado_jogador['Média_chute_Gol']=round(db_agrupado_jogador['Total_Chutes_Gol'] / db_agrupado_jogador['Aparicoes'],2)
db_agrupado_jogador['Média_gol']=round(db_agrupado_jogador['Total_Gols'] / db_agrupado_jogador['Aparicoes'],2)
db_agrupado_jogador['Taxa_Chutes_Gol'] = db_agrupado_jogador['Total_Chutes_Gol'] / db_agrupado_jogador['Total_Chutes']
```

```python
db_agrupado_time = database.groupby(['Time'], as_index=False).agg(
    Total_Gols=('Gols', 'sum'),
    Total_Chutes=('TC', 'sum'),   # Soma dos chutes ao gol
    Aparicoes=('Data', 'nunique'), # Contagem de jogos únicos (datas únicas)
    Total_Chutes_Gol=('CaG', 'sum') # Soma dos chutes ao gol
)
```

```python
db_agrupado_time['Média_chute']=round(db_agrupado_time['Total_Chutes'] / db_agrupado_time['Aparicoes'],2)
db_agrupado_time['Média_chute_gol']=round(db_agrupado_time['Total_Chutes_Gol'] / db_agrupado_time['Aparicoes'],2)
db_agrupado_time['Média_gol']=round(db_agrupado_time['Total_Gols'] / db_agrupado_time['Aparicoes'],2)
db_agrupado_time['Taxa_Chutes_Gol'] = db_agrupado_time['Total_Chutes_Gol'] / db_agrupado_time['Total_Chutes']
```

```python
db_agrupado_time[db_agrupado_time['Time'].isin(['Chelsea','West Ham United'])].sort_values(by='Total_Chutes', ascending=False, ignore_index= True)
```

```python
db_agrupado_time.sort_values(by='Média_chute_gol', ascending=False, ignore_index= True)
```

## Agrupado por Data

```python
db_desempenho_por_data = database.groupby(['Time', 'Data'], as_index=False).agg(
    Total_Gols=('Gols', 'sum'),          # Soma de gols por data
    Total_Chutes=('TC', 'sum'),          # Soma de chutes por data
    Total_Chutes_Gol=('CaG', 'sum')      # Soma dos chutes ao gol por data
).sort_values(by=['Time', 'Data'])
```

```python
db_desempenho_por_data['Taxa_Chutes'] = round(db_desempenho_por_data['Total_Gols'] / db_desempenho_por_data['Total_Chutes'],4)
db_desempenho_por_data['Taxa_Chutes_Gol'] = round(db_desempenho_por_data['Total_Gols'] / db_desempenho_por_data['Total_Chutes_Gol'],4)
```
